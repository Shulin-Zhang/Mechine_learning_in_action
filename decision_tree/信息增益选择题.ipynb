```python
from matplotlib import pylab as plt
from collections import Counter
import numpy as np

%matplotlib inline
```

```python
img = plt.imread('decision_choose.jpeg')
plt.figure(figsize=(10, 10))
plt.imshow(img)
```

```python
def calShannonEnt(Y):
    counter = Counter(Y)
    ent = 0
    total = len(Y)
    
    for key in counter.keys():
        prob = counter[key] / total
        ent -= prob * np.log2(prob)
        
    return ent
```

#### 1

```python
sunny = [1, 1, 0, 0, 0]
overcast = [1, 1, 1, 1]
rainy = [1, 1, 1, 0, 0]

Y1 = sunny + overcast + rainy
```

```python
Y1Ent = calShannonEnt(Y1)
```

```python
Y1SplitEnt = (len(sunny) / len(Y1)) * calShannonEnt(sunny) + (len(overcast) / len(Y1)) * calShannonEnt(overcast) \
                + (len(rainy) / len(Y1)) * calShannonEnt(rainy)
```

```python
Y1InfoGain = Y1Ent - Y1SplitEnt
Y1InfoGain
```
